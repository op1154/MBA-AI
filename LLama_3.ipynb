```python
from llama_cpp import Llama
```

```python
from huggingface_hub import hf_hub_download
```

```python
# Define the model ID and filename
model_id = "hugging-quants/Llama-3.2-3B-Instruct-Q4_K_M-GGUF"
model_filename = "llama-3.2-3b-instruct-q4_k_m.gguf"

# Download the model file
model_path = hf_hub_download(repo_id=model_id, filename=model_filename)

print(f"Model downloaded to: {model_path}")
```

Now, let's instantiate the Llama model using the downloaded `model_path`.

```python
# Instanciando o modelo
my_aweseome_llama_model = Llama(model_path=model_path)


prompt = "O que é inteligencia Artificial?"
max_tokens = 100
temperature = 0.3
top_p = 0.1
echo = True
stop = ["Q", "\n"]


# Define the parameters
model_output = my_aweseome_llama_model(
       prompt,
       max_tokens=max_tokens,
       temperature=temperature,
       top_p=top_p,
       echo=echo,
       stop=stop,
   )
final_result = model_output["choices"][0]["text"].strip()
```
