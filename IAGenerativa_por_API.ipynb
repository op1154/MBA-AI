```python
from google import genai

# Pegar dados de vari谩veis de ambiente
from google.colab import userdata

# SUBSTITUA pelo texto da sua chave que voc锚 copiou do AI Studio
MINHA_CHAVE_API = userdata.get('GOOGLE_API_KEY')

# Inicializa o cliente do Gemini
client = genai.Client(api_key=MINHA_CHAVE_API)
```

```python
# Escreva a sua pergunta aqui!
sua_pergunta = "Explique de forma simples o que 茅 uma intelig锚ncia artificial Generativa."

# Envia o pedido para o modelo mais r谩pido e eficiente (Gemini 2.5 Flash)
response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=sua_pergunta,
)

# Mostra a resposta gerada na tela
print("--- RESPOSTA DO GEMINI ---")
print(response.text)
```
